```python
import matplotlib.pyplot as plt

before_ai = [45, 50, 55, 60, 65]
after_ai = [60, 68, 72, 80, 90]

students = ["A", "B", "C", "D", "E"]

plt.plot(students, before_ai, label="Before AI", color="black")
plt.plot(students, after_ai, label="After AI", color="darkblue")

plt.xlabel("Students")
plt.ylabel("Grades")

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

before_ai = [45, 50, 55, 60, 65]
after_ai = [60, 68, 72, 80, 90]

students = ["A", "B", "C", "D", "E"]

x = np.arange(len(students))

plt.bar(x-0.2, before_ai, width=0.4, label="Before AI",color="black")
plt.bar(x+0.2, after_ai, width=0.4, label="After AI",color="grey")

plt.xticks(x, students)

plt.legend()

plt.show()
```
